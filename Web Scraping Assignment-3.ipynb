```python

driver = webdriver.Chrome()

# Define the categories
categories = ["fruits", "cars", "guitar","Machine Learning","cakes"]

for category in categories:
    # Open Google Images
    driver.get("https://images.google.com/")
    
    sign=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div/div/div/div/div[1]/div/div/div/a")
    sign.click()

    # Find and enter the search query
    search_input = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
    search_input.clear()
    search_input.send_keys(f"{category}")

    # Click the search button
    search_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span')
    search_button.click()

    # Wait for a short time to let the page load results
    time.sleep(2)

    # Scroll down to load more images (optional)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    # Extract image URLs
    category_images = []
    images = driver.find_elements(By.XPATH, '//div[@class="fR600b islir"]/img')

    for i in images:
        src_attribute = i.get_attribute("src")
        if src_attribute:
            category_images.append(src_attribute)

        if len(category_images) == 10:
            break

    # Print the URLs or save them to a file, database, etc.
    print(f"{category} images:")
    for image_url in category_images:
        print(image_url)

# Close the browser
driver.quit()
```

# 7.scrape details of billionaires from forbes.com 

```python
driver=webdriver.Chrome()
driver.get("https://forbes.com/")
```

```python
a=driver.find_element(By.XPATH,'//div[@class="card--large__title"]/a').click()
```

```python
Name=[]
name=driver.find_elements(By.XPATH,'//h2[@class="subhead-embed color-accent bg-base font-accent font-size text-align"]/sub/strong')
for i in data:
    Name.append(i.text)


    
        
```

```python
data=Name
```

```python
rank = []
name = []
source = []
networth = []

for entry in data:
    if entry[0].isdigit():
        # Split rank and name
        rank_name = entry.split('. ')
        rank.append(rank_name[0])
        name.append(rank_name[1])
    elif entry.startswith('Source'):
        # Extract source of wealth
        source.append(entry.split(': ')[1])
    elif entry.startswith('Net worth'):
        # Extract net worth
        networth.append(entry.split(': ')[1])

df=pd.DataFrame({"Rank":rank,"Name":name,"Source":source,"Net worth":networth})
df
```
