# 1.Scrape most viewed videos on youtube from wikipedia

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Name=[]
Artist=[]
Upload=[]
Views=[]
names=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in names:
    try:
        Name.append(i.text)
    except NoSuchElementException:
        Name.append("-")
artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist:
    try:
        Artist.append(i.text)
    except NoSuchElementException:
        Artist.append("-")
upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in upload:
    try:
        Upload.append(i.text)
    except NoSuchElementException:
        Upload.append("-")
views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in views:
    try:
        Views.append(i.text)
    except NoSuchElementException:
        Views.append("-")
```

```python
Rank=range(1,19)
Rank
```

```python
df=pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Artist,'Upload Date':Upload,'Views':Views})
df
```

# 7.scrape most watched Tv series of all time from imdb

```python
driver=webdriver.Chrome()
driver.get("https://www.imdb.com")
```

```python
menu=driver.find_element(By.XPATH,"/html/body/div[2]/nav/div[2]/label[1]").click()
TOP=driver.find_element(By.XPATH,"/html/body/div[2]/nav/div[2]/aside[1]/div/div[2]/div/div[2]/div[1]/span/div/div/ul/a[2]").click()
```

```python
URL=[]
url=driver.find_elements(By.XPATH,'//div[@class="sc-43986a27-0 gUQEVh cli-children"]/div/a')
for i in url:
    URL.append(i.get_attribute(('href')))
```

```python
Name=[]
YS=[]
Genre=[]
RT=[]
R=[]
V=[]
for a in URL:
    driver.get(a)
    time.sleep(5)
    try:
        name=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1')
        Name.append(name.text)
    except NoSuchElementException:
        Name.append("-")
    try:
        ys=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')
        YS.append(ys.text)
    except NoSuchElementException:
        YS.append("-")
    try:
        genre=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]')
        Genre.append(genre.text)
    except NoSuchElementException:
        Genre.append("-")
    try:
        rt=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[4]')
        RT.append(rt.text)
    except NoSuchElementException:
        RT.append("-")
    try:
        r=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')
        R.append(r.text)
    except NoSuchElementException:
        R.append("-")
    try:
        v=driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[3]')
        V.append(v.text)
    except NoSuchElementException:
        V.append("-")
        
        
```

```python
df=pd.DataFrame({"Name":Name,"Year Span":YS,"Genre":Genre,"Run Time":RT,"Ratings":R,"Votes":V})
df
```

```python
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
Datasets=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a').click()
```

```python
url=[]
start=0
end=10
for page in range(start,end):
    urls=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
    for i in urls:
        url.append(i.get_attribute(('href')))
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
```

```python
len(url)
```

```python
Name=[]
Type=[]
Task=[]
Atype=[]
NI=[]
NA=[]
Year=[]
for i in url:
    driver.get(i)
    time.sleep(5)
    try:
        name=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1')
        Name.append(name.text)
    except NoSuchElementException:
        Name.append("-")
    try:
        type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        Type.append(type.text)
    except NoSuchElementException:
        Type.append("-")
    try:
        task=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        Task.append(task.text)
    except NoSuchElementException:
        Task.append("-")
    try:
        atype=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        Atype.append(atype.text)
    except NoSuchElementException:
        Atype.append("-")
    try:
        ni=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
        NI.append(ni.text)
    except NoSuchElementException:
        NI.append("-")
    try:
        na=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
        NA.append(na.text)
    except NoSuchElementException:
        NA.append("-")
    try:
        year=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
        Year.append(year.text)
    except NoSuchElementException:
        Year.append("-")
```
